```python
import json
from deepface import DeepFace


# result=DeepFace.verify(img1_path="/content/rohit sharma.jpg",img2_path="/content/rohit.jpg")
# print(json.dumps(result,indent=2))

def detect_faces(image_path, db_path):
  dfs = DeepFace.find(img_path = image_path, db_path = db_path)
  data = dfs[0]
  imgs = data['identity'].tolist()
  return imgs


# dfs = DeepFace.find(img_path = "/content/rohit-sharma-4.jpg", db_path = "/content/database")
# print(dfs[0].split(''))
# data = dfs[0]
# imgs = data['identity'].tolist()

print(detect_faces('/content/rohit-sharma-4.jpg', '/content/database'))
```

```python
import json
import matplotlib.pyplot as plt
from deepface import DeepFace

# Load DeepFace model
# dfs = DeepFace.find(img_path="/content/rohit-sharma-4.jpg", db_path="/content/database")

# Get the images from the result
query_img = plt.imread("/content/rohit-sharma-4.jpg")
# found_img_paths = dfs['identity'].tolist()

found_imgs = [plt.imread(img_path) for img_path in detect_faces('/content/rohit-sharma-4.jpg', '/content/database')]

# Display the query image
plt.figure(figsize=(10, 5))
plt.subplot(1, len(found_imgs) + 1, 1)
plt.imshow(query_img)
plt.title('Query Image')
plt.axis('off')

# Display the found images
for i, img in enumerate(found_imgs):
    plt.subplot(1, len(found_imgs) + 1, i + 2)
    plt.imshow(img)
    plt.title(f"Found Image {i+1}")
    plt.axis('off')

plt.show()

# # Print the result
# print(json.dumps(dfs, indent=2))
```
